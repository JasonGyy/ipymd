```python
data = ipymd.data_input.crystal.Crystal(
    [[0.0, 0.0, 0.0], [0.5, 0.5, 0.5]], ['Na', 'Cl'], 
    225, cellpar=[.54, .54, .54, 90, 90, 90], 
    repetitions=[5, 5, 5])
```

```python
vis2 = ipymd.visualise_sim.Visualise_Sim()
vis2.add_box(*data.get_simulation_box())
vis2.add_axes()
atoms_df = data.get_atom_data()
vis2.add_atoms(atoms_df[atoms_df.type == 'Na'])
vis2.add_atoms(atoms_df[atoms_df.type == 'Cl'], alpha=0.2)
vis2.change_atom_radiimap({'Xx':0.1})
```

```python
images = [vis2.get_image(xrot=xrot,yrot=45) for xrot in [0,45,90]]
vis2.visualise(images, columns=2)
```

```python
import numpy as np

manip = ipymd.atom_manipulation.Atom_Manipulation()
data = ipymd.data_input.lammps.LAMMPS_Input(ipymd.get_test_path('crystal_data.data'))
vis3 = ipymd.visualise_sim.Visualise_Sim()
vis3.change_atom_radiimap({'Xx':0.1})
sim_abc, sim_origin = data.get_simulation_box()
#vis3.add_box(sim_abc,sim_origin,width=2)
vis3.add_axes(sim_abc, length=0.3,offset=(-0.4,-0.2),width=2)
atom_df = data.get_atom_data('charge')
repeat_df = manip.repeat_cell(atom_df,sim_abc,repetitions=((-2,2),(-2,2),(-1,1)))
repeat_df = manip.filter_variable(repeat_df, [1])
fe_origin=repeat_df.iloc[200][['xs','ys','zs']].values

#vis3.add_plane(3*sim_abc[0:2], sim_origin+np.array([0,0,7.4]), alpha=0.5, color='blue')

vis3.add_hexagon(sim_abc[[0,2]],fe_origin)

hex_df = manip.filter_inside_hexagon(repeat_df, sim_abc[[0,2]], fe_origin)
vis3.add_atoms(hex_df,type_map={1:'Fe',2:'S'})

img = vis3.get_image(xrot=0,yrot=0,fov=5,height=1000,width=1000)
#img.save('test.png')
vis3.visualise(img)
```

```python
vis3 = ipymd.visualise_sim.Visualise_Sim()
vis3.change_atom_radiimap({'Xx':0.1})

vis3.add_axes(sim_abc,length=0.5,offset=(-0.7,0.1))

repeat_df = manip.repeat_cell(atom_df,sim_abc,repetitions=((-1,2),(-1,2),(0,0)))
repeat_df = manip.change_variable(repeat_df, 1, 'Fe')
repeat_df = manip.change_variable(repeat_df, 2, 'S')

slice_df = manip.slice_z(repeat_df, 7, 10)

vis3.add_hexagon(sim_abc[[0,2]],sim_origin)

vis3.add_atoms(manip.filter_variable(slice_df, 'Fe'))
vis3.add_atoms(manip.filter_variable(slice_df, 'S'), alpha=0.4)
img1 = vis3.get_image(zrot=-30,fov=5)

s = 5.9
s2 = 1.4
slice_df2 = manip.slice_z(repeat_df, s-s2, s+s2)

vis3.remove_atoms(2)
vis3.add_atoms(manip.filter_variable(slice_df2, 'Fe'))
vis3.add_atoms(manip.filter_variable(slice_df2, 'S'), alpha=0.4)
img2 = vis3.get_image(zrot=-30,fov=5)

vis3.remove_atoms(2)
vis3.add_atoms(manip.filter_variable(repeat_df, 'Fe'))
vis3.add_atoms(manip.filter_variable(repeat_df, 'S'), alpha=0.3)
img3 = vis3.get_image(xrot=90,fov=5)

vis3.visualise([img2,img1,img3],columns=2)
```
